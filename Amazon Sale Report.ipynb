```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load and clean data
df = pd.read_csv(r"C:\Users\Lenovo\OneDrive\Desktop\New data anlayais project\Python_Amazon_Sales_Analysis-main\Amazon Sale Report.csv")
df.columns = df.columns.str.strip()

# Define custom color list with enough colors
custom_colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFD700',
                 '#FFCC99', '#CCFF66', '#FFB6C1', '#B0C4DE',
                 '#E6E6FA', '#FFA07A', '#98FB98']

# Create the countplot with hue fix
ax = sns.countplot(x='Size', hue='Size', data=df, palette=custom_colors, legend=False)

# Add bar labels
for bars in ax.containers:
    ax.bar_label(bars)

# Customize plot
plt.title("Count of Products by Size")
plt.xlabel("Product Size")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# note: From the above graph, you can see that most of the people buys M-Size

```python
#Group By:
#The Groupby() function in pandas is used to group data based on one or more columns in a DataFrame
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Group data
S_Qty = df.groupby('Size', as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)

# Plot
ax = sns.barplot(x='Size', y='Qty', hue='Size', data=S_Qty, palette='Set2', dodge=False, legend=False)

# Add labels
for bars in ax.containers:
    ax.bar_label(bars)

# Final touches
plt.title("Total Quantity Sold by Size")
plt.xlabel("Size")
plt.ylabel("Total Quantity")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(data=df,x='Courier Status',hue='Status')
```

Note: From above Graph the majority of the orders are shipped through the courner.

```python
#histogram
df['Size'].hist()
```

```python
df['Category']=df['Category'].astype(str)
column_data=df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=30,edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

```python
# Note: From above Graph You can see that most of the buyers are T-shirt
```

```python
#  Prepare data for pie chart
a1 = df['Fulfilment'].value_counts()

# Step 4: Plot the pie chart
fig, ax = plt.subplots()

ax.pie(a1, labels=a1.index, autopct='%1.1f%%', radius=0.7, wedgeprops=dict(width=0.6))
ax.set(aspect="equal")

plt.show()
```

```python
#prepare data for scatter plot
x_data = df['Category']
y_data = df['Size']
#plor the scatter plot
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
#plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution fo State')
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

category_sales = df.groupby('Category')['Amount'].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(8,6))
sns.barplot(x='Amount', y='Category', hue='Category', data=category_sales,
            palette='Set3', legend=False, dodge=False)
plt.title('Sales by Product Category')
plt.xlabel('Total Sales Amount')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
# top_10_States 
top_10_state = df['ship-state'].value_counts().head(10)
# Plot count of cities by state
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
```

```python
#note: From above graph you can see that most of the buyers are Maharashtra state
```
